```python
##problem statement.
## write a python program to calculate the gross salary of an employee based on basic salary.
#if basic<= 10000: HRA = 20%, DA= 80%
#IF BASIC<=20000: HRA = 25%, DA= 90%
#otherwise: HRA = 30%, DA = 95%
basic = float(input("Enter basic salary: "))

if basic <= 10000:
    hra = 0.20 * basic
    da = 0.80 * basic
elif basic <= 20000:
    hra = 0.25 * basic
    da = 0.90 * basic
else:
    hra = 0.30 * basic
    da = 0.95 * basic
gross_salary = basic + hra + da

print("Gross Salary =", gross_salary)




```
